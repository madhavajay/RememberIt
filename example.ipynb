```python
import rememberit

# rememberit.login("email@example.com", "password") # if you haven't logged in yet
decks = rememberit.sync()

def add(a, b):
    return a + b
    
deck_data = {
    "name": "Python Basics",
    "cards": [
        {
            "front": "How do you add two numbers in Python?",
            "front_type": "card",        # Styled gradient question card
            "front_theme": "purple",     # Theme: gradient/dark/light/blue/purple/green/orange
            "back": add,                 # Pass function directly! Source extracted automatically
            # back_type auto-detected as "code" when passing a function
        },
        {
            "front": "What is a list comprehension?",
            "front_type": "card",
            "front_theme": "blue",
            "back": "[x**2 for x in range(10) if x % 2 == 0]",
            "back_type": "code",
            "back_lang": "python",
        },
    ]
}
rememberit.upsert_deck(deck_data)
```
